```
dataEa = ea.loc[0]['DATA']
tipo = ea.loc[0]['TIPO']
resumo = ea.loc[0]['RESUMO DO EVENTO']
registro = ea.loc[0]['EVENTO']
sexo = ea.loc[0]['SEXO']
ID = ea.loc[0]['ID']
gravidade = ea.loc[0]['GRAVIDADE']

dataEa1 = ea.loc[1]['DATA']
tipo1 = ea.loc[1]['TIPO']
resumo1 = ea.loc[1]['RESUMO DO EVENTO']
registro1 = ea.loc[1]['EVENTO']
sexo1 = ea.loc[1]['SEXO']
ID1 = ea.loc[1]['ID']
gravidade1 = ea.loc[1]['GRAVIDADE']
```

```
with open('data.txt', 'w') as data:
  for i in range(42):
    dataEvol = evol.loc[i]['DATA EVOL']
    dadosEvol = evol.loc[i]['DADOS DA EVOLUÇÃO']
    data.write(str(i) + '\n')
    if(i < 16):
      data.write('Data EA: '+ str(dataEa) + '\n')
      data.write('Tipo: '+ tipo + '\n')
      data.write('Resumo: ' + resumo + '\n')
      data.write('Regsitro: ' + str(registro) + '\n')
      data.write('Sexo: ' + sexo + '\n')
      data.write('ID: ' + str(ID) + '\n')
      data.write('Gravidade: ' + gravidade + '\n')
    else:
      data.write('Data EA: '+ str(dataEa1) + '\n')
      data.write('Tipo: '+ tipo1 + '\n')
      data.write('Resumo: ' + str(resumo1) + '\n')
      data.write('Regsitro: ' + str(registro1) + '\n')
      data.write('Sexo: ' + sexo1 + '\n')
      data.write('ID: ' + str(ID) + '\n')
      data.write('Gravidade: ' + gravidade1 + '\n')
    data.write('\n')
    data.write('Data evolução: ' + dataEvol + '\n')
    data.write('Evolução: ' + str(dadosEvol))
    data.write(4*'\n')
    
with open('data.txt', 'r') as f2:
    data = f2.read()
    print(data)
  
  
```
